## Functions
Defined simply using ```def``` keyword. Can return as many variable as much you want. Can take optional keyword arguments as well.

````python
# simple python function
def poly(poly_idx, x):
    value = 0
    l = len(poly_idx) - 1
    
    for i in range(l+1):
        value += poly_idx[i]*(x**(l-i))
        
    return value

print("x^2 + x + 1 at x = 0 is:", poly([1, 1, 1], 0))
print("x^2 + x + 1 at x = 1 is:", poly([1, 1, 1], 1))
print("x^2 + x + 1 at x = 2 is:", poly([1, 1, 1], 2))
print("2x^2 + x + 1 at x = 2 is:", poly([2, 1, 1], 1))
````

````python
# function which can return multiple var
def get_data():
  return 1, 2, 3

x, y, z = get_data()

print(x, y, z)
````

### Try creating merge sort algorithm



Pseudo code is given below:

```
// SRC: https://www.geeksforgeeks.org/merge-sort/

MergeSort(arr[], l,  r)
If r > l
     1. Find the middle point to divide the array into two halves:  
             middle m = (l+r)/2
     2. Call mergeSort for first half:   
             Call mergeSort(arr, l, m)
     3. Call mergeSort for second half:
             Call mergeSort(arr, m+1, r)
     4. Merge the two halves sorted in step 2 and 3:
             Call merge(arr, l, m, r)
```

````python
def MergeSort(array):

  if len(array) > 1:
    # Finding the middle point to divide the array into two halves
    mid = len(array)//2

    # Finding the first half of the array
    First_Half = array[:mid]

    # Finding the second half of the array
    Second_Half = array[mid:]

    # Calling MergeSort() for the first half
    MergeSort(First_Half)

    # Calling MergeSort() for the second half
    MergeSort(Second_Half)

    # Algorithm to sort out the array elements
    i = 0
    j = 0
    k = 0

    while i < len(First_Half) and j < len(Second_Half):
      if First_Half[i] < Second_Half[j]:
        array[k] = First_Half[i]
        i += 1
      else:
        array[k] = Second_Half[j]
        j += 1
      k += 1

    # Checking if any element was left out
    while i < len(First_Half):
      array[k] = First_Half[i]
      i += 1
      k += 1

    while j < len(Second_Half):
      array[k] = Second_Half[j]
      j += 1
      k += 1

def print_Array(array):
  for i in array:
    print(i, end = " ")
  print()

def main():
  """
  Takes input from the user till the user presses Enter
  """
  array = []
  num = input("Enter a number to be added to the list which will be sorted :) - ")
  while num != "":
    array.append(int(num))
    num = input("Enter a number to be added to the list which will be sorted :) - ")

  print("Your list of numbers is :")
  print_Array(array)

  # Sorts the array
  MergeSort(array)
  print("The sorted array is :")
  print_Array(array)


if __name__ == "__main__":
  main()
````

### SciPy Stats
Those who don't have any backgrounds in statistics can get back here after completing the stats lecture.

````python
import scipy.stats

rv = scipy.stats.norm(0, 1) # getting a normal random var of mean=0 and var=1
````

````python
samples = rv.rvs(10000) # taking a random sample
print(samples)
print(samples.shape)
````

````python
x = np.arange(-4, 4, 0.05)
print(x.shape)
y = rv.pdf(x) #will give the probability distribution function of the random var 'rv'
print(y.shape)
````

````python
temp = plt.hist(samples, density=True, bins=100) # plot histogram i.e. frequency dist
plt.plot(x, y, 'r-', lw=3)
````
